```python
# ref: https://www.machinelearningplus.com/machine-learning/portfolio-optimization-python-example/

import numpy as np
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
payload=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
first_table = payload[0]
second_table = payload[1]

df = first_table
symbols = df['Symbol'].values.tolist()
print(len(symbols))
```

```python
# stocks = ['AAPL', 'GOOG', 'INTC', 'WFC', 'ADCP'] 
# stocks = yf.Tickers.download()
test = yf.download(symbols, period='1y')
# test = test.drop(columns = ['BF.B', 'BRK.B'])
test.head()
# Adj Close = L’adjusted close price o in italiano il prezzo di chiusura rettificato opera modificando il prezzo di chiusura di 
# un’azione per rifletterne il valore dopo aver tenuto conto di fattori quali dividendi, frazionamenti azionari e anche l’eventuale 
# emissione di nuove azioni. Viene spesso utilizzato quando si esaminano i rendimenti storici o si esegue un’analisi dettagliata 
# delle prestazioni passate
```

```python
test = test['Adj Close'] # prendo solo i prezzi di chiusura aggiustati
test = test.drop(columns = ['BF.B', 'BRK.B'])
test.head()
```

```python
# prendiamo il logaritmo del cambiamento percentuale giornaliero del valore delle azioni. si prende il logaritmo perché questo è
# tempo-additivo, qui p.e. presi i ritorni per tre periodi r12, r13 ed r23 posso dire che:
# log(r13) = log(r12) + log(r23)

# .pct_change() calcola il cambiamento percentuale => (x2-x1)/x1
log_apprezzamenti_quotidiani = test.pct_change().apply(lambda x: np.log(1+x))
log_apprezzamenti_quotidiani.head()
```

```python
# calcolo la varianza (campionaria, s**2), la dev standard non è altro che la radice quadrata di s**2. 
# MEMENTO: la dev standard attualmente è giornaliera, per avere quella del periodo occorre moltiplicarla per la shape[0] del ds

cov_matrix = test.pct_change().apply(lambda x: np.log(1+x)).cov()
cov_matrix
```

```python
corr_matrix = test.pct_change().apply(lambda x: np.log(1+x)).corr()
corr_matrix
```

##### la varianza del portafoglio si calcola così:
$$\sigma^2(Rp) = \sum{i=1}^{n} \sum_{j=1}^{n} w_i w_j COV(R_i, R_j) $$
##### dove wi e wj denotano i pesi degli asset da 1 a n, mentre Cov(Ri, Rj) le covarianze degli asset da 1 a n

```python
medie_diversi_er = log_apprezzamenti_quotidiani.mean()
medie_diversi_er 
```

```python
# # stabiliamo dei pesi casuali ed osserviamo l'expected return del portafoglio
# w = [0.1, 0.2, 0.5, 0.2] 
# port_er = (w*medie_diversi_er).sum()
# print('ER del portafoglio con pesi arbitrari: ' + str(port_er))
```

```python
# dev standard annuali dei diversi titoli. poiché si tratta di dev st annuali, moltipliachiamo per la rad quadrat del numero di 
# osservazioni (fonte: https://www.investopedia.com/articles/04/101304.asp#:~:text=Similarly%2C%20if%20we%20want%20to,trading%20days%20in%20a%20year).)
ann_sd = test.pct_change().apply(lambda x: np.log(1+x)).std().apply(lambda x: x*np.sqrt(test.shape[0]))
ann_sd
```

```python
# creiamo una tabella per misurare ritorni attesi e dev st degli asset
assets = pd.concat([medie_diversi_er, ann_sd], axis=1) # Creating a table for visualising returns and volatility of assets
assets.columns = ['Returns', 'Volatility']
assets
```

### Calcoliamo pesi casuali e cerchiamo il min variance portfolio

```python
# iterativamente cerchiamo i pesi. Poiché la somma dei pesi deve fare 1, dividiamo questi per la loro somma cumulata
import time

p_ret = [] # array dei ritorni del portfolio
p_vol = [] # array della volatilità del portfolio
p_weights = [] # array dei pesi degli asset
num_assets = len(test.columns)
num_portfolios = 500

for portfolio in range(num_portfolios):
    t = time.time()
    print('portfolio numero: ' + str(portfolio))
# for portfolio in range(1):
    weights = np.random.random(num_assets)
    weights = weights/np.sum(weights)
    p_weights.append(weights)
    # print('weights: ' + str(weights))
    # print('medie_diversi_er: ' + str(medie_diversi_er))
    # print('len di weights: ' + str(len(weights)))
    # print('shape di weights: ' + str(medie_diversi_er.shape))
    
    returns = np.dot(weights, medie_diversi_er) # Returns are the product of individual expected returns of asset and its 
                                      # weights 
    p_ret.append(returns)
    var = cov_matrix.mul(weights, axis=0).mul(weights, axis=1).sum().sum()# Portfolio Variance
    sd = np.sqrt(var) # Daily standard deviation
    ann_sd = sd*np.sqrt(250) # Annual standard deviation = volatility
    p_vol.append(ann_sd)
    print('Tempo per un singolo portfolio: ' + str(time.time() - t))
```

```python
# da: https://arxiv.org/pdf/1811.03975.pdf

# The quantum portfolio optimization algorithm can exhibit a run time that is logarithmic in the number
# of assets and the number of time steps. Suppose that there are N possible investments and T time instances, and our data set consists of T N-dimensional vectors of the historic returns for those investments
# stored in random access memory (RAM). Direct classical algorithms for portfolio optimization proceed by
# constructing the N ×N covariance matrix for the returns, using O

# T N2
# 
# operations/calls to RAM, and implementing its pseudo-inverse (O

# N2
# 
# operations/memory calls) to construct the optimal risk-return curve
# and to find the portfolio that maximizes return for a fixed value of ri
```

```python
data = {'Returns':p_ret, 'Volatility':p_vol}
for counter, symbol in enumerate(test.columns.tolist()):
    # print(counter, symbol)
    data[symbol +' weight'] = [w[counter] for w in p_weights]
```

```python
portfolios  = pd.DataFrame(data)
portfolios.head() # Dataframe of the 10000 portfolios created
```

```python
# plottiamo la frontiera efficiente

portfolios.plot.scatter(x='Volatility', y='Returns', marker='o', s=10, alpha=0.3, grid=True, figsize=[10,10])
plt.show()
```

```python
# vediamo il portfolio a varianza minima
min_vol_port = portfolios.iloc[portfolios['Volatility'].idxmin()]
print('portfolio a varianza minima: ')                             
print(min_vol_port)
# e plottiamolo
plt.subplots(figsize=[10,10])
plt.scatter(portfolios['Volatility'], portfolios['Returns'],marker='o', s=10, alpha=0.3)
plt.scatter(min_vol_port[1], min_vol_port[0], color='r', marker='*', s=500)
plt.show()
```

### Questa volta cerchiamo il miglior portfolio per Sharpe

```python
# scegliamo il portfolio adesso usando come criterio di scelta lo sharpe ratio. Questo è la differenza tra il rischio del portafoglio
# Rp ed il risk-free rate, cioè il ritorno di un investimento con rischio 0, cioè quello dell'investitore che non vuole prendersi
# alcun rischio (questo valore comunque non sarà mai 0). Questa differenza viene divisa per la dev st del portafoglio.
# Il miglior portafoglio rischioso è quello che massimizza lo Sharpe Ratio.

rf = 0.01 # stabiliamo un risk free pari a questa cifra
optimal_risky_port = portfolios.iloc[((portfolios['Returns']-rf)/portfolios['Volatility']).idxmax()]
print('Portfoglio che massimizza Sharpe:')
print(optimal_risky_port)
plt.subplots(figsize=(10, 10))
plt.scatter(portfolios['Volatility'], portfolios['Returns'],marker='o', s=10, alpha=0.3)
plt.scatter(min_vol_port[1], min_vol_port[0], color='r', marker='*', s=500)
plt.scatter(optimal_risky_port[1], optimal_risky_port[0], color='g', marker='*', s=500)
```
